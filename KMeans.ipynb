#### Kmeans is unsupervised machine learning method which helps to identify similar groups.

#### Kmeans claasify group cosidering distanace of data points  between groups and within group



#### Import Libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator  import make_blobs
from sklearn.cluster import KMeans # module for kmeans
```

#### Data Creation

```
x,y = make_blobs(n_samples=300, centers=4,cluster_std=0.6,random_state=0)
plt.scatter(x[:,0],x[:,1])
```

#### Cluster building

```
kmeans = KMeans(n_clusters=8, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(x)
plt.scatter(x[:,0], x[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```

##### Here, We can see that There are 8 clusters and they are not divided properly, In Kmeans We have to identify optimum no of clusters. For this we can use elbow method. In this method we assess the distnace of data points within clusters and between clusters.

it's better to have less distnce within cluster and more distance between clusters

#### Find optimum number of clusters

```
w = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', max_iter = 300, n_init=10, random_state=0 )
  kmeans.fit(x)
  w.append(kmeans.inertia_)
plt.plot(range(1,11), w)
plt.title('Elbow Method')
plt.xlabel('# of clusters')
plt.ylabel('W')
plt.show()
```

#### 4 is optimum no of clusters, Let's rerun the Model

```
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(x)
plt.scatter(x[:,0], x[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```

#### Now we can see cluster divide properly
